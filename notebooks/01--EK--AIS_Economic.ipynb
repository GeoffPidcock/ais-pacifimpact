### Fiji 

CPI Monthly

```python
# import data 

Fiji_CPI_raw = pd.read_csv('CPI_Fiji.csv')
Fiji_CPI_raw.head()
```

```python
#basic cleaning

# from columns 110 
Fiji_CPI_01 = Fiji_CPI_raw.iloc[:, np.r_[0,110:141]]

#keeping the Consumer price indices by COICOP, 2014  Base Year

Fiji_CPI_02 = Fiji_CPI_01.iloc[np.r_[1:15]]

#set index
Fiji_CPI_02 = Fiji_CPI_02.set_index('Descriptor')

#unstack 
Fiji_CPI_03 = Fiji_CPI_02.unstack().reset_index()

#name

Fiji_CPI_03['name'] = 'CPI_2014_' + Fiji_CPI_03['Descriptor']
Fiji_CPI_03['date'] = Fiji_CPI_03['level_0']
Fiji_CPI_03['value'] = Fiji_CPI_03[0]
```

```python
#second wrangling

Fiji_CPI_04 = Fiji_CPI_03

Fiji_CPI_04['frequency'] = 'monthly'
Fiji_CPI_04['country'] = 'fj'
Fiji_CPI_04['source'] = 'Official_Fiji'
Fiji_CPI_04['category'] = 'CPI'
Fiji_CPI_04['properties'] = '{currency = fjd}'

Fiji_CPI_04['metric_key'] = Fiji_CPI_04['name'] + '_' + Fiji_CPI_04['date']
```

```python
cols = ['metric_key','frequency','country','category','source','name','date','value','properties']
Fiji_CPI_04 = Fiji_CPI_04[cols]
Fiji_CPI_04.head()
```

### Samoa

GDP Quartely

```python
# import data

Samoa_GDP_raw = pd.read_csv('Samoa_GDP_quartely.csv')
Samoa_GDP_raw.head()
```

```python
#header

new_header = Samoa_GDP_raw.iloc[0]
Samoa_GDP_raw = Samoa_GDP_raw[1:] 
Samoa_GDP_raw.columns = new_header 


#remove rows 

Samoa_GDP_1 = Samoa_GDP_raw.iloc[np.r_[50:60]]
Samoa_GDP_1['date'] = Samoa_GDP_1.iloc[:,0]
Samoa_GDP_1 = Samoa_GDP_1.drop(Samoa_GDP_1.columns[0], axis=1)

#set index 
Samoa_GDP_1 = Samoa_GDP_1.set_index('date')

#unstack 
Samoa_GDP_2 = Samoa_GDP_1.unstack().reset_index(level = 'date')

#first level of wrangling 

Samoa_GDP_2['name'] = 'GDP_' + Samoa_GDP_2.index
Samoa_GDP_2['value'] = Samoa_GDP_2[0]
```

```python
#second wrangling 

Samoa_GDP_3 = Samoa_GDP_2

Samoa_GDP_3['frequency'] = 'quartely'
Samoa_GDP_3['country'] = 'wsm'
Samoa_GDP_3['source'] = 'Samoa_Bureau_Statistics'
Samoa_GDP_3['category'] = 'GDP'
Samoa_GDP_3['properties'] = '{currency = tala}'

Samoa_GDP_3['metric_key'] = Samoa_GDP_3['name'] + '_' + Samoa_GDP_3['date']
```

```python
#set order 

cols = ['metric_key','frequency','country','category','source','name','date','value','properties']
Samoa_GDP_3 = Samoa_GDP_3[cols]

#drop index
Samoa_GDP_3.reset_index( inplace=True)
Samoa_GDP_3 = Samoa_GDP_3.drop(Samoa_GDP_3.columns[0], axis=1)

Samoa_GDP_3.head()
```

```python
#second wrangling 

CK_GDP_3 = CK_GDP_2

CK_GDP_3['frequency'] = 'quartely'
CK_GDP_3['country'] = 'ck'
CK_GDP_3['source'] = 'National_stat_ck'
CK_GDP_3['category'] = 'GDP'
CK_GDP_3['properties'] = '{currency = nzd}'

CK_GDP_3['metric_key'] = CK_GDP_3['name'] + '_' + CK_GDP_3['date']
```

```python

cols = ['metric_key','frequency','country','category','source','name','date','value','properties']
CK_GDP_3 = CK_GDP_3[cols]
CK_GDP_3.head()
```

### Kiribati 

CPI quarterly

```python
# import data

Kiribati_CPI_raw = pd.read_csv('Kiribati_CPI_quart.csv')
```

```python
#clean

Kiribati_CPI_1 = Kiribati_CPI_raw

#remove columns 

Kiribati_CPI_1 = Kiribati_CPI_1.iloc[:, np.r_[1,147:171]]


#remove rows 

Kiribati_CPI_1 = Kiribati_CPI_1.iloc[np.r_[5:16]]
```

```python
#adding columns 

Kiribati_CPI_1.columns = ['industries',
    
                            '2018-01-01', 
                          '2018-02-01', 
                          '2018-03-01', 
                          '2018-04-01', 
                          '2018-05-01', 
                          '2018-06-01', 
                          '2018-07-01', 
                          '2018-08-01',
                          '2018-09-01', 
                          '2018-10-01', 
                          '2018-11-01', 
                          '2018-12-01', 
                          
                          '2019-01-01', 
                          '2019-02-01', 
                          '2019-03-01', 
                          '2019-04-01', 
                          '2019-05-01', 
                          '2019-06-01', 
                          '2019-07-01', 
                          '2019-08-01', 
                          '2019-09-01', 
                          '2019-10-01', 
                          '2019-11-01', 
                          '2019-12-01']
```

```python
#unstack 


Kiribati_CPI_2 = Kiribati_CPI_1.set_index('industries')
Kiribati_CPI_2 = Kiribati_CPI_2.unstack().reset_index()
```

```python
#wrangling

Kiribati_CPI_2['name'] = 'CPI_2006_' + Kiribati_CPI_2['industries']
Kiribati_CPI_2['value'] = Kiribati_CPI_2[0]


#columns and final wrangling 

Kiribati_CPI_3 = Kiribati_CPI_2

Kiribati_CPI_3['frequency'] = 'monthly'
Kiribati_CPI_3['country'] = 'kir'
Kiribati_CPI_3['source'] = 'Official_kiribati'
Kiribati_CPI_3['category'] = 'CPI'
Kiribati_CPI_3['properties'] = 'none'
Kiribati_CPI_3['date'] = Kiribati_CPI_3['level_0'] 

Kiribati_CPI_3['metric_key'] = Kiribati_CPI_3['name'] + '_' + Kiribati_CPI_3['date']


#set order 

cols = ['metric_key','frequency','country','category','source','name','date','value','properties']
Kiribati_CPI_3 = Kiribati_CPI_3[cols]
Kiribati_CPI_3.head()
```
