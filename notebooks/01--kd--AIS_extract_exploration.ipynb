```python
#change variable types where necessary
AIS_extract_fiji['dtg'] = pd.to_datetime(AIS_extract_fiji['dtg'])
AIS_extract_fiji[['mmsi','vessel_type_code','flag_code']] = AIS_extract_fiji[['mmsi','vessel_type_code','flag_code']].apply(lambda x: x.astype(str))
```

### Filtering

```python
#remove duplicates
AIS_extract_fiji.drop_duplicates(keep = False, inplace = True)
```

```python
#check only those at anchor or moored
AIS_fiji_at_port = AIS_extract_fiji[(AIS_extract_fiji['nav_status'] == 'At Anchor')| (AIS_extract_fiji['nav_status'] == 'Moored')]
```

```python
#check if enough datapoints for further filtering
AIS_fiji_at_port['vessel_type'].value_counts()
```

```python
#only cargo and tanker necessary for merchandise trade
AIS_fiji_mt = AIS_fiji_at_port[(AIS_fiji_at_port['vessel_type'] == 'Cargo')| (AIS_fiji_at_port['vessel_type'] == 'Tanker')]
```

```python
#plot of count of unique vessels per month
AIS_fiji_mt.set_index('dtg').resample('M').nunique()['mmsi'].plot()
```

```python
#keep only values with date
AIS_fiji_mt_dff = AIS_fiji_mt[AIS_fiji_mt['dtg'].notnull()]
```

```python
AIS_output_testing = AIS_fiji_mt_dff[AIS_fiji_mt_dff['flag_country'] != 'Fiji'][['dtg', 'vessel_name']]
```

Checking if ships visit more than once per month

```python
#Unique ships for plotting
all_ships = AIS_output_testing['vessel_name'].unique().tolist()
```

```python
#check only 20 records
sample_ships = all_ships[:20]
```

```python
#datetime as index column
AIS_output_testing.set_index('dtg',inplace=True)
```

```python
#creating a small multiples plot to see all data at once
fig, axes = plt.subplots(nrows=4, ncols=5, sharey=True, figsize=(10,10))
axes_list = [item for sublist in axes for item in sublist]

for ship in sample_ships:
    
    ax = axes_list.pop(0)
    AIS_output_testing[AIS_output_testing['vessel_name']==ship].resample('D').nunique().plot(ax=ax)
    ax.tick_params(
        which='both',
        bottom='off',
        left='off',
        right='off',
        top='off'
    )
    ax.grid(linewidth=0.25)
    ax.set_xlabel("")
    ax.spines['left'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_legend().remove()
plt.tight_layout()
```

Noticed some ships enter port multiple times per month

```python
#Creating unique identifier for each day
AIS_fiji_mt_dff['day'] = AIS_fiji_mt_dff[AIS_fiji_mt_dff['flag_country'] != 'Fiji']['dtg'].dt.to_period('D')
```

```python
#cleaning to only keep minimum columns necessary
AIS_fiji_mt_cleaned = AIS_fiji_mt_dff[AIS_fiji_mt_dff['flag_country'] != 'Fiji'][['day', 'vessel_name']]
```

```python
#removing duplicates to keep 1 record per vessel per day
AIS_fiji_mt_unique = AIS_fiji_mt_cleaned.drop_duplicates()
```

```python
AIS_fiji_mt_unique.sort_values(by=['vessel_name','day'])
```

```python
#have to add back timestamp to be able to calculate difference in records
AIS_fiji_mt_unique['day'] = AIS_fiji_mt_unique['day'].dt.to_timestamp()
```

```python
#calculate number of days between previous record. If difference is not 1, then ship entered/re-entered the port
AIS_fiji_mt_unique['diff'] = AIS_fiji_mt_unique.groupby('vessel_name')['Day'].diff() / np.timedelta64(1, 'D')
AIS_fiji_mt_unique['diff'] = AIS_fiji_mt_unique['diff'].fillna(0)
```

```python
#sorting data
AIS_fiji_mt_unique = AIS_fiji_mt_unique.sort_values(by=['vessel_name','day'])
```
