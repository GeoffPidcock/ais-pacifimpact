```python
#read in raw file, ideally directly from DB
AIS_extract = pd.read_csv('lautoka-fiji.csv', error_bad_lines=False)
```

```python
#change variable types where necessary
AIS_extract['dtg'] = pd.to_datetime(AIS_extract['dtg'])
AIS_extract[['mmsi','vessel_type_code','flag_code']] = AIS_extract[['mmsi','vessel_type_code','flag_code']].apply(lambda x: x.astype(str))
```

```python
#remove duplicates
AIS_extract.drop_duplicates(keep = False, inplace = True)
```

```python
#check only those at anchor or moored
AIS_at_port = AIS_extract[(AIS_extract['nav_status'] == 'At Anchor')| (AIS_extract['nav_status'] == 'Moored')]
```

```python
#check unique types if ships
AIS_at_port['vessel_type'].unique()
```

```python
#Only tanker and cargo seem relevant for merchandise trade
AIS_mt = AIS_at_port[(AIS_at_port['vessel_type'] == 'Cargo')| (AIS_at_port['vessel_type'] == 'Tanker')]
```

```python
#Creating unique identifier for each day
AIS_mt['day'] = AIS_mt['dtg'].dt.to_period('D')
```

```python
#have to add back timestamp to conduct analysis
AIS_mt['day'] = AIS_mt['day'].dt.to_timestamp()
```

```python
#sort_values
AIS_mt.sort_values(by=['mmsi','dtg'],inplace=True)
```

```python
AIS_mt.drop_duplicates(subset=['mmsi','day'],inplace=True)
```

```python
AIS_mt['entryid']=AIS_mt.day.diff().dt.days.ne(1).cumsum()
```

```python
AIS_mt.set_index('dtg').resample('M').nunique()['entryid'].plot()
```

```python
import_test_fiji = pd.read_csv(r'C:\Users\karan\Google Drive\Civita\AIS\hackathon data\notebooks\test_import.csv')
```

```python
#convert  datetime 
import_test_fiji['month'] = pd.to_datetime(import_test_fiji['date']).dt.to_period('M')
```

```python
AIS_mt_monthly = AIS_mt.set_index('day').resample('M').nunique()
```

```python
import_test_fiji_merged.set_index('date',inplace=True)
```

```python
fig, ax_left = plt.subplots(figsize=(15,8))
ax_right = ax_left.twinx()

ax_left.plot(import_test_fiji_merged['value'].astype(float), color='black')
ax_right.plot(import_test_fiji_merged['mmsi'].astype(float), color='red')
plt.tight_layout()
```
